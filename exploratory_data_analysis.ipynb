```python
# for exploratory data analysis, importing the requeired libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib inline is a magic command in Python that allows Jupyter Notebook or JupyterLab to 
# display plots and graphs directly in the notebook interface
```

```python
df = pd.read_csv('automobileEDA.csv')
```

```python
df.head()
```

```python
#Analyzing individual features pattern using visualization

#listing the data type for each of the column
print(df.dtypes)
```

```python
#checking the data type of single column named peak-rpm
df['peak-rpm'].dtypes
```

```python
# we can calculate the correlation between variables of type "int64" or "float64" using the method "corr":
df.corr()
```

```python
# Find the correlation between the following columns: bore, stroke, compression-ratio, and horsepower.
df[['bore','stroke','compression-ratio','horsepower']].corr()
```

```python
# finding the scatter plot of engine size and price
#engine size as potential predictor variable of price
sns.regplot(x="engine-size", y = "price", data=df)
plt.ylim(0,)
```

```python
# As the engine-size goes up, the price goes up: this indicates a positive direct 
# correlation between these two variables. Engine size seems like a pretty good 
# predictor of price since the regression line is almost a perfect diagonal line.
```

```python
# We can examine the correlation between 'engine-size' and 'price' and see that it's approximately 0.87.
df[["engine-size", "price"]].corr()
```

```python
# highway mpg is a potential predictor variable of price and finding the scatterplot of "highway-mpg" and "price"
sns.regplot(x = "highway-mpg", y="price", data=df)

# As highway-mpg goes up, the price goes down: this indicates an inverse/negative 
# relationship between these two variables. Highway mpg could potentially be a predictor of price.
```

```python
# finding the correlation between highway-mpg and price
df[['highway-mpg', 'price']].corr()
```
